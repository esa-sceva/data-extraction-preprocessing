```python
from gliner import GLiNER
model = GLiNER.from_pretrained("E3-JSI/gliner-multi-pii-domains-v1").to('cuda')
```

```python
print(model.device)
```

```python
import os
from tqdm.auto import tqdm

labels = ["name", "organizations", "phone number", "email", "email address", "paper id", "orcid id"]

metadata = {}
for _f in tqdm(os.listdir('/content/drive/MyDrive/pii_extraction_sample')):
    with open(f"/content/drive/MyDrive/pii_extraction_sample/{_f}", 'r') as f:
      text = f.read()

      local_ents = {}
      splits = [text[i:i + 300] for i in range(0, len(text), 300)]
      for sentence in splits: # doing thus to avoid max length truncation(384), beyond that perf is bad
        entities = model.predict_entities(sentence, labels, threshold = 0.5)
        for entity in entities:
          local_ents[ entity["text"] ] = entity["label"]

      metadata[_f] = local_ents
```
