```python
from docling.document_converter import DocumentConverter
import os
from tqdm.auto import tqdm

DATA_PATH = '/content/'
OUTPUT_PATH = './output/'

for _file in tqdm(os.listdir(DATA_PATH)):
  if _file.endswith('.pdf'):
    converter = DocumentConverter()
    result = converter.convert(os.path.join(DATA_PATH, _file))
    with open(f"{_file.replace('.pdf', '.md')}", 'w', encoding = 'utf-8') as f:
      f.write(result.document.export_to_markdown())
```
