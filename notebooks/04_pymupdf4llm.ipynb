```python
import pymupdf4llm
import os
from tqdm.auto import tqdm
import pathlib

DATA_PATH = '../data/extraction_metrics/files/'
OUTPUT_PATH = '../data/extraction_metrics/predictions/pypdf4llm/'

for _file in tqdm(os.listdir(DATA_PATH)):
    md_text = pymupdf4llm.to_markdown(os.path.join(DATA_PATH, _file))
    pathlib.Path(os.path.join(OUTPUT_PATH, _file.replace('.pdf', '.md'))).write_bytes(md_text.encode())
```
