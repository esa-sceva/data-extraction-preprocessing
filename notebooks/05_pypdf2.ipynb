```python
from PyPDF2 import PdfReader
import os
from utils import save_to_markdown
from tqdm.auto import tqdm

DATA_PATH = '../data/extraction_metrics/files/'
OUTPUT_PATH = '../data/extraction_metrics/predictions/pypdf2/'


for _file in tqdm(os.listdir(DATA_PATH)):
    with open(os.path.join(DATA_PATH, _file), 'rb') as file:
        reader = PdfReader(file)
        text = ""
        for page in reader.pages:
            page_text = page.extract_text()
            if page_text:
                text += page_text

    save_to_markdown(text, os.path.join(OUTPUT_PATH, _file.replace('.pdf', '.md')))
```
