```python
from unstructured.partition.auto import partition
from tqdm.auto import tqdm
import os

DATA_PATH = '/content/'
OUTPUT_PATH = './output/'

for _file in tqdm(os.listdir(DATA_PATH)):
  try:
    elements = partition(os.path.join(DATA_PATH, _file))
    with open(f"{_file.replace('.pdf', '.md')}", "w", encoding="utf-8") as f:
          f.write("\n\n".join([str(el) for el in elements]))
  except:
    pass
```
