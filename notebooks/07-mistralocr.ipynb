```python
MISTRAL_KEY = 'xxx'
```

```python
from mistralai import Mistral
import os
from utils import save_to_markdown
from tqdm.auto import tqdm

DATA_PATH = '../data/files/'
OUTPUT_PATH = '../data/predictions/mistral/'

api_key = MISTRAL_KEY
client = Mistral(api_key=api_key)

for _file in tqdm(os.listdir(DATA_PATH)):
    uploaded_pdf = client.files.upload(
        file={
            "file_name": f"{_file}",
            "content": open(os.path.join(DATA_PATH, _file), "rb"),
        },
        purpose = "ocr"
    )  

    signed_url = client.files.get_signed_url(file_id = uploaded_pdf.id)

    ocr_response = client.ocr.process(
        model = "mistral-ocr-latest",
        document={
            "type": "document_url",
            "document_url": signed_url.url,
        }
    )

    text = ""
    for i in range(len(ocr_response.pages)):
        text += ocr_response.pages[i].markdown
    
    save_to_markdown(text, os.path.join(OUTPUT_PATH, _file.replace('.pdf', '.md')))
```
