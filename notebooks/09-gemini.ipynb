```python
GEMINI_KEY = ''

import google.generativeai as genai
genai.configure(api_key = GEMINI_KEY)

model = genai.GenerativeModel('gemini-2.0-flash')

from PIL import Image
import os
from tqdm.auto import tqdm

FOLDER = "../data/images/-AIST-2D---Results-of-flight-tests-and-applica_2023_The-Egyptian-Journal-of-"

def save_to_markdown(text, path):
    with open(path, "w", encoding="utf-8") as f:
        f.write(text)
    
    print(f"saved to {path}")

def ocr_with_gemini(image_paths):
    """Process images with Gemini 2.0 Flash for OCR"""
    images = [Image.open(os.path.join(FOLDER, path)) for path in tqdm(os.listdir(image_paths))]
    
    prompt = """
    Extract all the text, tables and formulas in a markdown format.
    """
    
    response = model.generate_content([prompt, *images])
    return response.text

text = ocr_with_gemini(FOLDER)
save_to_markdown(text, f'../data/predictions/gemini/{FOLDER.split('/')[-1]}' + '.md')

```
