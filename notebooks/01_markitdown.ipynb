```python
from markitdown import MarkItDown
import os

from utils import save_to_markdown
from tqdm.auto import tqdm

DATA_PATH = '../data/extraction_metrics/files/'
OUTPUT_PATH = '../data/extraction_metrics/predictions/markitdown/'

for _file in tqdm(os.listdir(DATA_PATH)):
    md = MarkItDown()
    result = md.convert(os.path.join(DATA_PATH, _file))
    save_to_markdown(result.text_content, os.path.join(OUTPUT_PATH, _file.replace('.pdf', '.md')))
```
