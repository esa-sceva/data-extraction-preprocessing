```python
from pdf2image import convert_from_path
import os

for _file in os.listdir():
  if _file.endswith('.pdf'):
    pages = convert_from_path(f'/content/{_file}')

    os.makedirs(f'./data/{_file.split(".")[0]}')

    for count, page in enumerate(pages):
      page.save(f'./data/{_file.split(".")[0]}/{count}.jpg', 'JPEG')
```
